## Data Wrangling For Top 50 Crypto

### Set Up

```python
#Imports
import os
import pandas as pd
import numpy as np

# Functions
import sys
sys.path.append('./helpers')
from helpers import load_data_csv_files
```

```python
# Data Loading
# Load all 49 cryptocurrency datasets from raw data directory
# Each CSV contains historical OHLC data and volume information
crypto_data = load_data_csv_files('../data/raw/')

# Access individual datasets like this:
# aave_df = crypto_data['aave']
# algorand_df = crypto_data['algorand']
# etc.
print(f"\nTotal cryptocurrencies loaded: {len(crypto_data)}")
```

```python
crypto_data.keys()
```

```python
# Print Shape for all Data sets
for crypto_name, df in crypto_data.items():
    print(f"{crypto_name}: {df.shape}")
```

```python
for crypto_name, df in crypto_data.items():
    print(f"{crypto_name}: {df.head()}")
```

```python
# Remove the first row of each DataFrame
# The raw data includes a metadata/header row that duplicates column information
# This row must be removed before further processing
for crypto_name, df in crypto_data.items():
    crypto_data[crypto_name] = df.iloc[1:]
    
print("First row removed from all datasets")
```

```python
for crypto_name, df in crypto_data.items():
    print(f"{crypto_name}: {df.iloc[0]}")
```

### Missing Data

```python
# Find and remove missing data
# Strategy: Rows with any missing values are removed to ensure data quality
# This approach is suitable for time-series data where completeness is important
for crypto_name, df in crypto_data.items():
    missing_count = df.isnull().sum()
    if missing_count.sum() > 0:
        print(f"{crypto_name} Missing Data Before Cleaning:\n{missing_count}\n")
        rows_before = len(df)
        crypto_data[crypto_name] = df.dropna()
        rows_after = len(crypto_data[crypto_name])
        print(f"  -> Removed {rows_before - rows_after} rows with missing values\n")
    else:
        print(f"{crypto_name}: No missing values found\n")
```

```python
# Value counts 
for crypto_name, df in crypto_data.items():
    print(f"{crypto_name} Value Counts:\n{df['Open'].value_counts()}\n")  # Replace 'column_name' with the actual column you want to analyze
    print(f"{crypto_name} Value Counts:\n{df['Close'].value_counts()}\n")
    print(f"{crypto_name} Value Counts:\n{df['Volume'].value_counts()}\n")
```

### Change Data Types

```python
# Convert data types for consistency across all cryptocurrency datasets
# This ensures that all 50 datasets have uniform types regardless of source variation
for crypto_name, df in crypto_data.items():
    # Convert Date to datetime object for proper time-series handling
    if 'Date' in df.columns:
        df['Date'] = pd.to_datetime(df['Date'])
    
    # Convert OHLC (Open, High, Low, Close) prices to float for calculations
    if 'Close' in df.columns:
        df['Close'] = pd.to_numeric(df['Close'], errors='coerce')
    if 'Open' in df.columns:
        df['Open'] = pd.to_numeric(df['Open'], errors='coerce')
    if 'High' in df.columns:
        df['High'] = pd.to_numeric(df['High'], errors='coerce')
    if 'Low' in df.columns:
        df['Low'] = pd.to_numeric(df['Low'], errors='coerce')
    
    # Convert Volume to numeric (can be int or float) for analysis
    if 'Volume' in df.columns:
        df['Volume'] = pd.to_numeric(df['Volume'], errors='coerce')
    
    # Update the dataframe in the dictionary
    crypto_data[crypto_name] = df
```

```python
# Dtypes
for crypto_name, df in crypto_data.items():
    print(f"{crypto_name} Data Types:\n{df.dtypes}\n")
```

### Save Data

```python
# Save the Cleaned DataFrames back to CSV files
cleaned_data_dir = '../data/interim/'
for crypto_name, df in crypto_data.items():
    cleaned_file_path = os.path.join(cleaned_data_dir, f"{crypto_name}_cleaned.csv")
    df.to_csv(cleaned_file_path, index=False)
    print(f"Saved cleaned data for {crypto_name} to {cleaned_file_path}")
```
